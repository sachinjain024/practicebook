```python
import graphlab
```

```python
song_data = graphlab.SFrame('song_data.gl/')
```

```python
song_data.head()
```

```python
unique_songs = song_data['song_id'].unique()
len(unique_songs)
```

```python
unique_users = song_data['user_id'].unique()
len(unique_users)
```

```python
graphlab.canvas.set_target('ipynb')
song_data['title'].show()
```

## Create a popularity based recommender

```python
train_data, test_data = song_data.random_split(0.8, seed = 0)
```

```python
popularity_model = graphlab.popularity_recommender.create(train_data, user_id='user_id', item_id='song')
```

```python
popularity_model.recommend(users = [unique_users[0]])
```

```python
popularity_model.recommend(users = [unique_users[5]])
```

> Conclusion: See the problem with popularity recommender that it recommends the same things to every user.

## Build song recommender with personalization

```python
personalized_model = graphlab.item_similarity_recommender.create(
    train_data,
    user_id='user_id',
    item_id='song'
)
```

```python
users = unique_users
personalized_model.recommend(users=[users[0]])
```

```python
personalized_model.recommend(users=[users[2]])
```

> Note that now we get different recommendations for different users

```python
personalized_model.get_similar_items(items=['With Or Without You - U2'])
```

```python
personalized_model.get_similar_items(['Chan Chan (Live) - Buena Vista Social Club'])
```

### Question3: let’s use to find the song that is most recommended by the personalized_model model

```python
train_data, test_data = song_data.random_split(0.8, seed=0)
item_similarity_model = graphlab.item_similarity_recommender.create(train_data, user_id='user_id', item_id='song')
```

```python
subset_test_users = test_data['user_id'].unique()[0:10000]
len(subset_test_users)
```

```python
recommendations = item_similarity_model.recommend(users=subset_test_users, k=1)
print recommendations
```

```python
recommendations_grouped = recommendations.groupby(key_columns='song', operations={'count': graphlab.aggregate.COUNT()})
print recommendations_grouped
```

```python
recommendations_sorted = recommendations_grouped.sort('count', ascending=False)
```
