```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
PATH = '../data/'
```

```python
df_train = pd.read_csv(f'{PATH}train.csv', index_col='PassengerId')
df_test = pd.read_csv(f'{PATH}test.csv', index_col='PassengerId')
```

```python
df_train.head()
```

```python
df_test.head()
```

## Fill Missing Values

```python
target = df_train['Survived']
df_train = df_train.drop('Survived', axis=1)
df_train['training_set'] = True
df_test['training_set'] = False
df_full = pd.concat([df_train, df_test])
df_full = df_full.interpolate()
```

```python
## get_dummies does one-hot encoding
df_full = pd.get_dummies(df_full)

## Recreate df_train and df_test
df_train = df_full[df_full['training_set']==True]
df_train = df_train.drop('training_set', axis=1)
df_test = df_full[df_full['training_set']==False]
df_test = df_test.drop('training_set', axis=1)
```

## Create Model

```python
rfc = RandomForestClassifier()
rfc_model = rfc.fit(df_train, target)

## Check the score
rfc_model.score(df_train, target)
```

```python
## Predictions
predictions = rfc.predict(df_test)
```

```python
## Submissions
my_submission = pd.DataFrame({'PassengerId': df_test.index, 'Survived': predictions})
my_submission.to_csv(f'{PATH}submission_random_forest_classifer.csv', index=False)
```
